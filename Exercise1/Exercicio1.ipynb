Import das Dependencias e Declarção das imagens

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img1 = cv2.imread("image3-1.png" , cv2.COLOR_BGR2GRAY)
img2 = cv2.imread("image3-2.png" , cv2.COLOR_BGR2GRAY)
img3 = cv2.imread("image3-3.png" , cv2.COLOR_BGR2GRAY)

Lmax = 255

def printImg(img):
  cv2_imshow(img)
  #ou :
  # plt.imshow(img)
  # plt.show()
  #para exibir na grade com plt


print("Inputs: ")
printImg(img1)
printImg(img2)
printImg(img3)
```

Transformação de potência com diferentes valores de gama.

```python
def powerTransform(img,Lmax,gama):
  img = img.astype(float)                     #convertendo toda a imagem de uint8 para float( a fim de facilitar nas contas)

  C = Lmax/ np.power(img.max(),gama)          #definindo a constante C para ser aplicada em cada pixel da imagem

  for i in range(len(img)):
    for j in range(len(img[i])):
      img[i][j]= C*np.power(img[i][j],gama)  #formula para transformação em potencia

  img = np.uint8((img / np.max(img)) * 255)   #retornando os valores para uint8. Format utilizado para "imprimir" a imagem

  return img

#GAMA < 1: realça trechos escuros || GAMA > 1 : realça trechos claros
GAMA = 2.3
img1_power = powerTransform(img1,Lmax,GAMA)
img2_power = powerTransform(img2,Lmax,GAMA)
img3_power = powerTransform(img3,Lmax,GAMA)

GAMA1 = 0.5
img4_power = powerTransform(img1,Lmax,GAMA1)
img5_power = powerTransform(img2,Lmax,GAMA1)
img6_power = powerTransform(img3,Lmax,GAMA1)

print("Transformação de Potência: ")
print(f"gama = {GAMA}")
printImg(img1_power)
printImg(img2_power)
printImg(img3_power)
print(f"gama = {GAMA1}")
printImg(img4_power)
printImg(img5_power)
printImg(img6_power)
```

Transformação Logaritma

```python
def logaritimTransform(img,Lmax):
  img = img.astype(float)                     #convertendo toda a imagem de uint8 para float( a fim de facilitar nas contas)

  C = Lmax/np.log(1+img1.max())               #definindo a constante C para ser aplicada em cada pixel da imagem

  for i in range(len(img)):
    for j in range(len(img[i])):
      img[i][j]= C*(np.log(1+img[i][j]))      #formula para transformação logaritma

  img = np.uint8((img / np.max(img)) * 255)   #retornando os valores para uint8. Format utilizado para "imprimir" a imagem

  return img

img1_log = logaritimTransform(img1,Lmax)
img2_log = logaritimTransform(img2,Lmax)
img3_log = logaritimTransform(img3,Lmax)

print("Transformação Logaritma: ")
printImg(img1_log)
printImg(img2_log)
printImg(img3_log)




```
